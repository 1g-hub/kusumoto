```python
import numpy as np
import math

# グラフプロット用基準関数
def f(x1, x2):
    fx = (x1-1) ** 2 + 10 * (x1**2-x2) ** 2
    return fx
# 基準関数の微分
def df(x1, x2):
    dx1fx = (40*x1)**3 - (40*x1*x2) + 2*x1 -2
    dx2fx = 20*x2 - (20*x1)**2
    dx = np.array([dx1fx, dx2fx])
    return dx

# 勾配降下法に必要なパラメータ
eta = 100                         # 学習率
max_iteration = 529                # 最大反復回数
x1 = 0                          # 初期値x0
x2 = 1                             # 初期値y0
x_pred = [x1]                       # 描画用x0軌跡リスト(初期値をプリセット)
y_pred = [x2]                       # 描画用y0軌跡リスト(初期値をプリセット)


# 最大反復回数まで計算する

for i in range(100, 1, -1):
    eta = i/1000
    x1=0
    x2=1

    for j in range(max_iteration):
      x1, x2 = np.array([x1, x2]) - eta * df(x1, x2)          # 勾配降下法
      x_pred.append(x1)               # x0の軌跡をリストに追加
      y_pred.append(x2)               # y0の軌跡をリストに追加
      #print(j, x1, x2)
      a1, a2 =  df(x1, x2)
      norm = math.sqrt(a1**2 + a2**2)

      if norm < 1/1000:
        break
```
